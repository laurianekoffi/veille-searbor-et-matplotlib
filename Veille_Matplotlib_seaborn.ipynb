# III) REGRESSION LOGISTIQUE

```python
data
```

```python
#Encodage des variables qualitatives
from sklearn.preprocessing import LabelEncoder

lb_make = LabelEncoder()
df = data
for i in quali:
    df[i] = lb_make.fit_transform(data[i])
    
df
```

```python
#Diviser le dataframe
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df, df['coeur'], test_size=0.3, random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.preprocessing import StandardScaler
```

```python
#Mise à l'échelle des variables explicatives
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
#Création du modèle et entrainement
model_log = LogisticRegression(random_state = 0, solver='liblinear')
model_log.fit(X_train,y_train)
```

```python
#Prédiction
y_pred = model_log.predict(X_test)
y_pred
```

```python
#Evaluation (Performance du modèle)
performance = model_log.score(X_test,y_test)
performance
```

```python
#Matrice de confusion
M_confusion = confusion_matrix(y_test, y_pred)
M_confusion
```

```python
print(classification_report(y_test, y_pred))
```
